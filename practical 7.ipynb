```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
documentA 	= 	'Jupiter is the largest Planet'
documentB 	= 	'Mars is the fourth planet from the Sun'
```

```python
documents 	= 	[documentA, documentB]
```

```python
stop_words	 =	 set(stopwords.words('english'))
```

```python
processed_docs = []

for doc in documents:
    doc = doc.lower()
    words = word_tokenize(doc)
    filtered_words = [w for w in words if w not in stop_words]
    processed_docs.append(filtered_words)

print("Processed Documents:")

for i, doc in enumerate(processed_docs):
    print(f"Doc {i+1}: {doc}")
```

```python
vocab = set()
for doc in processed_docs:
    vocab.update(doc)
vocab = list(vocab)
print("\nVocabulary:", vocab)
```

```python
def compute_tf(doc):
    tf_dict = {}
    doc_len = len(doc)
    print("\nTF Calculation:")
    for word in vocab:
        count = doc.count(word)
        tf = count / doc_len
        tf_dict[word] = tf
        print(f"{word} = {count}/{doc_len} = {tf}")

    return tf_dict
tf_docs = []
for i, doc in enumerate(processed_docs):
    print(f"\n--- TF for Document {i+1} ---")
    tf_docs.append(compute_tf(doc))
```

```python
from collections import Counter
docA = "this is a sample document this is test"
bagOfWordsA = docA.split()
wordDictA = Counter(bagOfWordsA)
numOfWordsA = len(bagOfWordsA)
print("Bag of Words:", bagOfWordsA)
print("Word Count:", wordDictA)
print("Total Words:", numOfWordsA)
```

```python
import math
def compute_idf(docs):
    N = len(docs)
    idf_dict = {}
    print("\nIDF Calculation:")
    for word in vocab:
        count = sum(1 for doc in docs if word in doc)
        idf = math.log(N / (count + 1)) + 1   # smoothing
        idf_dict[word] = idf
        print(f"{word} → appears in {count} docs → IDF = {idf}")
    return idf_dict
idf = compute_idf(processed_docs)
```
